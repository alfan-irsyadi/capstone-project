```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ori_df = pd.read_csv('MariBisnis.csv')
df = ori_df
df.head()
```

```python
df.date = pd.to_datetime(df.date)
df.yr_built = pd.to_datetime(df.yr_built, format='%Y')
df['age']=(df.date - df.yr_built)/np.timedelta64(1,'Y')
df.set_index('id', inplace=True)
df.head()
```

```python
df = df.select_dtypes(np.number)
df
```

```python
df.boxplot(figsize=(15,10))
plt.show()
```

```python
from scipy.stats import zscore
z_scores = zscore(df)
filter = (np.abs(z_scores)<3).all(axis=1)
df = df[filter]
df.head()
```

```python
df.boxplot(figsize=(15,10))
plt.show()
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
df
```

```python
df.hist(bins=30,figsize=(15,10))
plt.show()
```

```python
X = df.drop(['price', 'lat', 'long'], axis=1)
y = df.price
X.boxplot(figsize=(15,10))
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
import warnings
warnings.filterwarnings('ignore')
select = SelectKBest(score_func=f_classif, k=10);
select = select.fit(X,y)
kolom = select.get_support(indices=True)
X = X.iloc[:,kolom]
X
```

```python
df.quantil(0.75)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.22, random_state = 10)
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR_fit = LR.fit(X_train, y_train)
y_prediction = LR_fit.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
score=r2_score(y_test,y_prediction)
print('r2 socre is ',score)
print('mean_sqrd_error is==',mean_squared_error(y_test,y_prediction))
print('root_mean_squared error of is==',np.sqrt(mean_squared_error(y_test,y_prediction)))
y_pred = LR_fit.predict(X)
```

```python
y = pd.DataFrame(y)
y['y_pred'] = y_pred
```
