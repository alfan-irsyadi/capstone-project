```python
to_drop = ['id', 'long', 'lat', 'zipcode', 'yr_renovated', 'date']
data_dropped = data.drop(to_drop, axis=1)
```

```python
df = data.copy()
df.date = pd.to_datetime(df.date)
df.yr_built = pd.to_datetime(df.yr_built, format='%Y')
df['age']=(df.date - df.yr_built)/np.timedelta64(1,'Y')
df.head()
```

```python
df['is_renovated'] = np.where(df['yr_renovated']==0, df['yr_renovated'], 1)
df.head()
```

```python
df['has_basement'] = np.where(df['sqft_basement']==0, df['sqft_basement'], 1)
df.head()
```

```python
arr = []
kondisi = df['id'].value_counts()>1
kondisi_index = np.where(kondisi.values, arr.append(kondisi.index),0)
arr
```

```python
kategori = ['is_renovated','has_basement' ,'date', 'condition', 
            'zipcode', 'grade', 'view', 'waterfront', 'yr_built', 'yr_renovated', 'id']
numerik = df.drop(kategori, axis = 1).columns
df[numerik]
```

```python
from scipy.stats import zscore
z_scores = zscore(df[numerik])
filter = (np.abs(z_scores)<3).all(axis=1)
df = df[filter].reset_index(drop=True)
```
